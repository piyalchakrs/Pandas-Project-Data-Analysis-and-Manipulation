# Classify movies based on ratings [Excellent,Good and Average]

```python
data.columns
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else: 
        return "Average"
```

```python
data['Rating_Catergory'] = data['Rating'].apply(rating)
data.head()
```

# Find unique values from genre

```python
data.columns
```

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
unique_list=[]
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
```

```python
unique_list
```

```python
len(unique_list)
```
